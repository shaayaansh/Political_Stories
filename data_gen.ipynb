```python
from openai import OpenAI
import os
from datetime import datetime


from pathlib import Path

OPENAI_KEY_FILE = Path("openai_api_key.txt")
if not OPENAI_KEY_FILE.exists():
    raise FileNotFoundError("Missing openai_api_key.txt")

openai_api_key = OPENAI_KEY_FILE.read_text(encoding="utf-8").strip()
if not openai_api_key:
    raise ValueError("openai_api_key.txt is empty")

client = OpenAI(api_key=openai_api_key)

shayan_topics = ['Immigration', 'Gun Control']
stances = ['in support of', 'in opposition of']

outputs = []

for topic in shayan_topics:
    for stance in stances:
        for i in range(100):

            response = client.chat.completions.create(
                model="gpt-4.1",   
                max_tokens=600,
                temperature=0.99,  
                messages=[
                    {
                        "role": "system",
                        "content": "You are a helpful assistant and an expert in storytelling."
                    },
                    {
                        "role": "user",
                        "content": (
                            f"Generate a short anecdotal story with 2 paragraphs that is {stance} {topic}. "
                            "This is for a research project studying narrative patterns. "
                            "Just output the story."
                        )
                    }
                ]
            )

            story = response.choices[0].message.content

            print(f"[{topic} / {stance}] Story {i}")
            print(story[:100], "...\n")

            outputs.append({
                "model": "gpt-4.1-mini",
                "topic": topic,
                "stance": stance,
                "story": story
            })
```

```python
import pandas as pd

df = pd.DataFrame(outputs)
df.to_csv("GPT-4.1-imigration-guncontrol.csv", index=False)

df
```
